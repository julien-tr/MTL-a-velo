# Collisions routières


```python
df = pd.read_csv(path+"collisions_routieres/collisions_routieres.csv",skipinitialspace=True)
```

```python
df.head()
```

```python
print(df.shape)
print(df.dtypes)
print(df.info())
#print(df.describe())
```

```python
# Keep only related columns

df = df[['DT_ACCDN','JR_SEMN_ACCDN','HEURE_ACCDN','REG_ADM','VITESSE_AUTOR','CD_GENRE_ACCDN','CD_ETAT_SURFC','CD_ECLRM','CD_CONFG_ROUTE','CD_COND_METEO','LOC_LAT','LOC_LONG','NB_DECES_VELO','NB_BLESSES_VELO','NB_VICTIMES_VELO']]
df
```

```python
# Keep only related rows
df = df[(df['NB_DECES_VELO'] == 1) | (df['NB_BLESSES_VELO'] == 1) | (df['NB_VICTIMES_VELO'] == 1)]
df
```

```python
# Create features : year, month, day, hour,  Daytime - Nighttime. 


# Convert to datetime
df.DT_ACCDN = pd.to_datetime(df.DT_ACCDN, format='%Y-%m-%d', errors='coerce')

df
```

```python
df['year'] = pd.DatetimeIndex(df['DT_ACCDN']).year
df['month'] = pd.DatetimeIndex(df['DT_ACCDN']).month
df['week-day'] = pd.DatetimeIndex(df['DT_ACCDN']).weekday
df.head()
```

```python
df
```

```python
# Building the map
# Load an empty map
from keplergl import KeplerGl
map_1 = KeplerGl()
map_1
```

```python
map_1.add_data(data=df, name='velol')
```
