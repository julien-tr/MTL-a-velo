```python
# Keep only related columns

df = df[['DT_ACCDN','JR_SEMN_ACCDN','HEURE_ACCDN','REG_ADM','VITESSE_AUTOR','CD_GENRE_ACCDN','CD_ETAT_SURFC','CD_ECLRM','CD_CONFG_ROUTE','CD_COND_METEO','LOC_LAT','LOC_LONG','NB_DECES_VELO','NB_BLESSES_VELO','NB_VICTIMES_VELO']]
df
```

```python
# Keep only related rows
df = df[(df['NB_DECES_VELO'] == 1) | (df['NB_BLESSES_VELO'] == 1) | (df['NB_VICTIMES_VELO'] == 1)].copy()
```

```python
df.HEURE_ACCDN.value_counts()
```

```python
# Create features : year, month, day, hour,  Daytime - Nighttime. 


# Convert to datetime
df.DT_ACCDN = pd.to_datetime(df.DT_ACCDN, format='%Y-%m-%d', errors='coerce')
df['HEURE_ACCDN'] = [x[0:5] for x in df['HEURE_ACCDN']]
df['HEURE_ACCDN'] = df['HEURE_ACCDN'].replace('NA',)
df
```
